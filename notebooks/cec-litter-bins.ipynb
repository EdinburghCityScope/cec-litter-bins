```python
import folium;
import json;

geoJson=json.load(open('../data/litter-bins.geojson'))

litter_bins = folium.Map(location=[55.9450167, -3.1894458],
                    zoom_start=13)

folium.GeoJson(geoJson,
               name='geojson'
              ).add_to(litter_bins)

folium.LayerControl().add_to(litter_bins)
litter_bins
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import json
import  requests
import numpy as np
from collections import Counter

# Load the data into a geoJson object
geoJson=json.load(open('../data/litter-bins.geojson'))

# Count the number of areas
c = Counter(feature['properties']['type'] for feature in geoJson['features'])
sizes = []
values = []
explode = []
for count in c:
    if (count!='general') and (count!='NULL'):
        #Add the values, sizes, and explode if over 50%
        values.append(count)
        sizes.append(c[count])

fig = plt.figure()
ax = plt.subplot(111)
ax.bar(range(len(values)), sizes, width=1)
ax.set_xticks(np.arange(len(values))+1)
ax.set_xticklabels(values,rotation=90)
#ax.xaxis_date()
print("Number of bins by type")
```
